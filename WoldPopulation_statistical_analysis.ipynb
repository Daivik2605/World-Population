### Lets create a function that lets us get the required statistical information of our target columns
#### What we need:
- Min
- Max
- Count
- Range
- Mean
- Median
- Mode
- Q1
- Q3
- Quartile Upper Limit
- Quartile Lower Limit
- Variance
- Standard Deviation


#### target columns --> 2022 Population, Area

```python
population = df['2022 Population']
area = df['Area (km²)']
```

```python
def statistical_analysis(column, column_name):
    print(f"\nStatistical Analysis for {column_name}:")

    # Count
    count = column.count()
    print(f"Count: {count}")

    # Minimum and Maximum
    min_val = column.min()
    max_val = column.max()
    print(f"Minimum: {min_val}")
    print(f"Maximum: {max_val}")

    # Range (Max - Min)
    range_val = max_val - min_val
    print(f"Range: {range_val}")

    # Interquartile Range (IQR)
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    print(f"Interquartile Range (IQR): {IQR}")

    # Lower and Upper Limits for Outliers
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    print(f"Lower Limit (Q1 - 1.5*IQR): {lower_limit}")
    print(f"Upper Limit (Q3 + 1.5*IQR): {upper_limit}")

    # Median
    median = column.median()
    print(f"Median: {median}")

    # Mean
    mean = column.mean()
    print(f"Mean: {mean}")

    # Mode
    mode = column.mode()
    if mode.empty:
        print("Mode: NA")
    else:
        print(f"Mode: {mode[0]}")  # Takes the first mode in case of multiple modes

    # 1st and 3rd Quartiles
    print(f"1st Quartile (Q1): {Q1}")
    print(f"3rd Quartile (Q3): {Q3}")

    # Sample Variance and Sample Standard Deviation
    variance = column.var()
    std_dev = column.std()
    print(f"Sample Variance: {variance}")
    print(f"Sample Standard Deviation: {std_dev}")

    # Outliers
    outliers = column[(column < lower_limit) | (column > upper_limit)]
    print(f"Outliers: {outliers.values}")
```

```python
statistical_analysis(df['2022 Population'], "2022 Population")
```

```python
statistical_analysis(df['Area (km²)'], "Area (km²)")
```

### Standard Deviation Plot

```python
mean_area = df['Area (km²)'].mean()
std_dev_area = df['Area (km²)'].std()
```

```python
import matplotlib.pyplot as plt
# Plot histogram with KDE (Kernel Density Estimate)
plt.figure(figsize=(8, 6))
sns.histplot(df['Area (km²)'], kde=True, color='skyblue', bins=10)

# Add mean and standard deviation lines
plt.axvline(mean_area, color='red', linestyle='--', label=f'Mean: {mean_area:.2f}')
plt.axvline(mean_area + std_dev_area, color='green', linestyle='--', label=f'+1 SD: {mean_area + std_dev_area:.2f}')
plt.axvline(mean_area - std_dev_area, color='green', linestyle='--', label=f'-1 SD: {mean_area - std_dev_area:.2f}')

# Add title and labels
plt.title('Standard Deviation Plot of Area')
plt.xlabel('Area (km²)')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.show()
```
